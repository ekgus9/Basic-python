```python
# 코랩에는 matplolib 설치됨
import matplotlib as mpl
import matplotlib.pyplot as plt

plt.style.use('classic') # 기본 스타일 설정
```

```python
# 인터페이스

# matlab 스타일: 단순, 편리, 사용자 친화적
# 객체 지향 스타일: 복잡하고 상세한 설정, 파이썬 프로그램 친화적
```

1. 선그림(Line plot)

```python
# y = f(x) 시각화

import numpy as np

plt.show() # 선 보이게 함

fig = plt.figure() # 그림판
fig.suptitle("Line Plot") # 제목
ax = plt.axes() # 축

x = np.arange(1,10)
plt.plot(x,np.sin(x),color = 'blue',linestyle = '--') # :k 과 같이 검정색과 선스타일 간단히 지정 가능
plt.xlim(11,-1) # 범위
plt.axis([-1,11,-12,12]) # xy축 범위
plt.yscale("log") # 축의 스케일 변경
plt.title("Line plot")
plt.xlabel("x")
```

```python
plt.plot(x,np.sin(x),'-g',label = 'sin(x') # 범례
plt.legend() # 범례 표시
```

2. 점그림(Scatter plot)

```python
# 산점도

x=np.linspace(0,10,25)
plt.plot(x,x,'o',color='r')
plt.plot(x,x+2,'.',color='g')
plt.plot(x,x+4,'s-b',) # 사각형
plt.plot(x,x+6,'d--',color='c') # 마름모
```

```python
plt.scatter(x,x,alpha=0.4) # 산점도, alpha는 투명도
```

3. 다중 서브 플롯

```python
plt.style.use('seaborn-whitegrid')

nrow, ncol = 3, 4
for i in range(nrow*ncol):
  plt.subplot(nrow,ncol,i+1)
  plt.text(0.5,0.5,str((nrow,ncol,i+1)),fontsize=16,ha='center')
```

```python
grid = plt.GridSpec(3,4,wspace=0.4,hspace=0.4)
ax1 = plt.subplot(grid[0,0])
ax2 = plt.subplot(grid[0,1:3])
```

```python
# 정규분포
mean = [0,0]
cov = [[1,1],[1,2]]
x,y= np.random.multivariate_normal(mean,cov,5000).T

fig=plt.figure(figsize=(6,6))
grid=plt.GridSpec(4,4)
main_ax=fig.add_subplot(grid[:-1,1:])
y_hist=fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=main_ax)

main_ax.plot(x,y,'ok',markersize=3,alpha=0.2)
y_hist.hist(y,40,histtype='stepfilled',orientation='horizontal',color = 'gray')
y_hist.invert_xaxis()
```
